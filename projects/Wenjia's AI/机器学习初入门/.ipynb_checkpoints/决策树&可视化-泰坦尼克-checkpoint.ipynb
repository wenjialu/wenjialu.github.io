```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import DictVectorizer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_graphviz
```

```python
import pandas as pd
titan = pd.read_csv("/Users/lujiawen/Desktop/数据分析课-note/数据集/titanic.txt")
titan.head()
```

### 数据处理

```python
x = titan[["pclass", "age", "sex"]]
y = titan["survived"]
```

```python
np.any(pd.isnull(x['pclass']))
np.any(pd.isnull(x['age']))
```

```python
# 缺失值需要处理，将特征当中有类别的这些特征进行字典特征抽取
x['age'].fillna(x['age'].mean(), inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=22)
```

```python
x_train.head()
```

### 特征工程

#### 字典

```python
x_train.to_dict()

x_train.to_dict(orient="records")  #orient="records"???
```

```python
transfer = DictVectorizer(sparse=False)

x_train = transfer.fit_transform(x_train.to_dict(orient="records"))
x_test = transfer.transform(x_test.to_dict(orient="records"))
x_train
```

### 训练-决策树

```python
# 4.机器学习(决策树)
estimator = DecisionTreeClassifier(criterion="entropy", max_depth=5)
estimator.fit(x_train, y_train)

# 5.模型评估
estimator.score(x_test, y_test)

estimator.predict(x_test)
```

### 可视化

```python
export_graphviz(estimator, out_file="/Users/lujiawen/Desktop/数据分析课-note/机器学习/titan-tree.dot", feature_names=['age', 'pclass=1st', 'pclass=2nd', 'pclass=3rd', '女性', '男性'])
```
