```python
df = df[['分区', 'author','date','coins','danmu','favorite','likes','replay','share','view']]
df.head()
```

```python
sc = df.loc[df['分区'] == '科学科普']
so= df.loc[df['分区']=='社科人文']
ma = df.loc[df['分区']=='机械']
tec = df.loc[df['分区']=='野生技术协会']
mi = df.loc[df['分区']=='星海'] # 一般发布军事内容
car = df.loc[df['分区']=='汽车']
sc
```

```python
count = sc.groupby('author')['view'].count().reset_index() #每个up的视频数，取任意一个column
count.columns = ['author','times'] #重命名
#总时长
last = sc.groupby('author')['date'].max()  #每个up最新发布时间
late = sc.groupby('author')['date'].min() #每个up最早发布时间、
# a = last- late
```

```python
#计算每个up发布间隔天数 = 总时长/视频数 ？

fre = round((last-late).dt.days/sc.groupby('author')['view'].count()).reset_index()
fre.columns = ['author', 'Fre']

# 剔除掉发布视频少于5的up主
up_big = count[count['times']>5]


new_column = pd.merge(up_big,fre,on='author',how='inner')

new_column
```
