```python
#nltk.download('punkt') 不安装会报错
default_tagger = nltk.DefaultTagger("NN") #默认tagger会标注出现频率最大的词性（如，名词）
raw= "嗷嗷 喵 嘞个咪"
tokens= nltk.word_tokenize(raw)
tags = default_tagger.tag(tokens) 
print(tags)
```

```python
tagger = nltk.RegexpTagger([(r'.*们','PRO')]) #把带"们"的设置为代词
print(tagger.tag(nltk.word_tokenize('我们 和 你们啊')))
```
