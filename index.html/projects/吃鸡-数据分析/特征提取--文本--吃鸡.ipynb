```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('/Users/lujiawen/Desktop/PUBG_to_stu/data/test_V2.csv')
data.head()
```

```python
data['matchType'].head()
```

```python
matchType= data['matchType']
transfer = CountVectorizer()
# 调用fit_transform
mt = transfer.fit_transform(matchType)
print("文本特征抽取的结果：\n", mt.toarray())
print("返回特征名字：\n", transfer.get_feature_names())  #只能是比赛类型中的1个，所以只有一个1.
```

```python
data['matchType'] = mt.toarray()
data['matchType'].sort_value() 
```
